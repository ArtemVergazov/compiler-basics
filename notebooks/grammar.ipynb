```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
grammar = nx.DiGraph(incoming_graph_data={
    'Prog': {
        'Stmt': {'label': 'stmts*'},
    },
    'Stmt': {
        'StmtLet': {'label': 'stmt'},
        'StmtExit': {'label': 'stmt'},
        'StmtIf': {'label': 'stmt'},
        'Scope': {'label': 'stmt'},
    },
    'StmtLet': {
        'Token': {'label': 'identifier'},
        'Expr': {'label': 'expr'},
    },
    'StmtExit': {
        'Expr': {'label': 'expr'},
    },
    'StmtIf': {
        'BranchIf': {'label': 'ifBranch'},
        'BranchElif': {'label': 'elifBranches*'},
        'BranchElse': {'label': 'elseBranch'},
    },
    'Scope': {
        'Stmt': {'label': 'stmts*'},
    },
    'Expr': {
        'Term': {'label': 'expr'},
        'BinExpr': {'label': 'expr'},
    },
    'BranchIf': {
        'Expr': {'label': 'expr'},
        'Scope': {'label': 'scope'},
    },
    'BranchElif': {
        'Expr': {'label': 'expr'},
        'Scope': {'label': 'scope'},
    },
    'BranchElse': {
        'Scope': {'label': 'scope'},
    },
    'BinExpr': {
        'BinExprAdd': {'label': 'expr'},
        'BinExprSub': {'label': 'expr'},
        'BinExprMul': {'label': 'expr'},
        'BinExprDiv': {'label': 'expr'},
    },
    'Term': {
        'TermParen': {'label': 'term'},
        'TermIntLiteral': {'label': 'term'},
        'TermIdentifier': {'label': 'term'},
    },
    'BinExprAdd': {
        'Expr': {'label': 'lhs'},
        'Expr': {'label': 'rhs'},
    },
    'BinExprSub': {
        'Expr': {'label': 'lhs'},
        'Expr': {'label': 'rhs'},
    },
    'BinExprMul': {
        'Expr': {'label': 'lhs'},
        'Expr': {'label': 'rhs'},
    },
    'BinExprDiv': {
        'Expr': {'label': 'lhs'},
        'Expr': {'label': 'rhs'},
    },
    'TermParen': {
        'Expr': {'label': 'expr'},
    },
    'TermIntLiteral': {
        'Token': {'label': 'intLiteral'},
    },
    'TermIdentifier': {
        'Token': {'label': 'identifier'},
    },
})
```

```python
plt.figure(figsize=(20, 10))
pos = nx.kamada_kawai_layout(grammar)
nx.draw_networkx(
    grammar,
    pos=pos,
    node_size=100,
    # font_size=8,
)
nx.draw_networkx_edge_labels(
    grammar, pos,
    edge_labels={(n1, n2): grammar[n1][n2]['label'] for n1, n2 in grammar.edges},
    # font_size=8,
);
```
